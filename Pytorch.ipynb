```python
# BTVN 1: Tính đạo hàm của y = 5x^5 + 6x^3 - 3x + 1 tại x = 2
x = torch.tensor(2.0, requires_grad=True)
y = 5*x**5 + 6*x**3 - 3*x + 1
y.backward()
print("Đạo hàm tại x=2:", x.grad)
```

# BTVN 2:
Tạo một tensor ban đầu có giá trị là 2
Định nghĩa hàm số và tính gradient
y = x^3 2x^2 + 5x + 1
hãy tính dy/dx tại giá trị của x
Dùng phương pháp Gradient Descent với Learning rate alpha = 0.1 để cập nhật giá trị x trong 10 vòng lặp

```python
# BTVN 2: Gradient Descent cho y = x^3 + 2*x^2 + 5*x + 1
x = torch.tensor(2.0, requires_grad=True)
alpha = 0.1
for i in range(10):
    y = x**3 + 2*x**2 + 5*x + 1
    y.backward()
    print(f"Vòng {i+1}: x = {x.item():.4f}, dy/dx = {x.grad.item():.4f}")
    with torch.no_grad():
        x -= alpha * x.grad
    x.grad.zero_()
```

# BTVN 3:
Tạo một tập dữ liệu giả lập với x là số giờ học (ngẫu nhiên từ 1 - 10) và y là số điểm được tính theo công thức y = 3x + 5 + noise
Với noise là một giá trị ngẫu nhiên nhỏ
1. Khởi tạo tham số w và b ngẫu nhiên với requires_grad = True
2. Tính MSE
3. Tính Gradient
4. Cập nhật tham số w và b bằng Gradient Descent với Learning rate alpha = 0.01
5. Lặp lại quá trình trên trong 100 vòng lặp để xem sự hội tụ của mô hình

```python
# BTVN 3: Linear Regression với Gradient Descent
import torch
import numpy as np

# Tạo dữ liệu giả lập
np.random.seed(42)
x_data = np.random.uniform(1, 10, 100)
noise = np.random.normal(0, 1, 100)
y_data = 3 * x_data + 5 + noise

x_tensor = torch.tensor(x_data, dtype=torch.float32).view(-1, 1)
y_tensor = torch.tensor(y_data, dtype=torch.float32).view(-1, 1)

# Khởi tạo tham số
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

alpha = 0.01
epochs = 100

for epoch in range(epochs):
    # Dự đoán
    y_pred = w * x_tensor + b
    
    # MSE
    loss = torch.mean((y_pred - y_tensor)**2)
    
    # Gradient
    loss.backward()
    
    # Cập nhật
    with torch.no_grad():
        w -= alpha * w.grad
        b -= alpha * b.grad
    
    # Reset grad
    w.grad.zero_()
    b.grad.zero_()
    
    if (epoch+1) % 10 == 0:
        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f}, w = {w.item():.4f}, b = {b.item():.4f}")

print("Final: w ≈ 3, b ≈ 5")
```

```python
# BTVN 5: Tạo tensor với empty, zeros, ones, random và reshape
import torch

# Empty tensor
empty_tensor = torch.empty(3, 3)
print("Empty tensor:")
print(empty_tensor)

# Zeros tensor
zeros_tensor = torch.zeros(3, 3)
print("\nZeros tensor:")
print(zeros_tensor)

# Ones tensor
ones_tensor = torch.ones(3, 3)
print("\nOnes tensor:")
print(ones_tensor)

# Random tensor
random_tensor = torch.rand(3, 3)
print("\nRandom tensor:")
print(random_tensor)

# Reshape với view
reshaped_view = random_tensor.view(9)
print("\nReshaped with view (1D):")
print(reshaped_view)

# Reshape với view_as (giả sử có tensor khác)
other_tensor = torch.zeros(1, 9)
reshaped_view_as = random_tensor.view_as(other_tensor)
print("\nReshaped with view_as (1x9):")
print(reshaped_view_as)
```
